```python
def display_board(x):
    print('\n')
    print(f' {x[1]} | {x[2]} | {x[3]}')
    print(f'-----------')
    print(f' {x[4]} | {x[5]} | {x[6]}')
    print(f'-----------')
    print(f' {x[7]} | {x[8]} | {x[9]}')
    print('\n')
```

```python
def clean_board():
    return ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
```

```python
def legend_board():
    return ['#','1','2','3','4','5','6','7','8','9']
```

```python
def player_input():
    p1_marker='null'
    while p1_marker!='O' and p1_marker!='X':
        p1_marker=input('Player 1: Chose your mark X or O: ')
        if p1_marker!='O' and p1_marker!='X':
            print('You did not chose correcly')
        elif p1_marker=='O':
            p2_marker='X'
        elif p1_marker=='X':
            p2_marker='O'
    a=[0,p1_marker,p2_marker]
    return a
```

```python
def place_marker(board, marker, position):
    board[position]=marker
```

```python
def win_check(board, mark):
    if board[1]==board[2]==board[3]==mark:
        return True
    elif board[4]==board[5]==board[6]==mark:
        return True
    elif board[7]==board[8]==board[9]==mark:
        return True
    elif board[1]==board[4]==board[7]==mark:
        return True
    elif board[2]==board[5]==board[8]==mark:
        return True
    elif board[3]==board[6]==board[9]==mark:
        return True
    elif board[1]==board[5]==board[9]==mark:
        return True
    elif board[3]==board[5]==board[7]==mark:
        return True
    else:
        pass
```

```python
from random import randint

def choose_first():
    return randint(1,2)
```

```python
def space_check(board, position):
    if board[position]==' ':
        return True
    else:
        return False
```

```python
def full_board_check(board):
    if ' ' in board:
        return True
    else:
        return False
```

```python
def player_choice(board):
    available=False
    while available==False:
        a=input('Where to put your sign?: ')
        if a.isdigit()==False:
            print('It is not a number!')
        elif int(a) not in range(1,10):
            print('It is not a number between 1-9')
        else:
            a_int=int(a)
            available=space_check(board,a_int)
            if available==False:
                print('Position already taken!')
    return a_int
```

```python
def switch(x):
    if x == 1:
        return 2
    elif x == 2:
        return 1
```

```python
board=clean_board()
legend=legend_board()
mark=player_input()
display_board(legend)
game_on=True
turn=choose_first()

while game_on==True:

    print(f'Player {turn} turn:')
    position=player_choice(board)
    place_marker(board,mark[turn],position)
    display_board(board)
  
    if win_check(board,mark[1])==True:
        game_on=False
        print('Player 1 has won!')

    elif win_check(board,mark[2])==True:
        game_on=False
        print('Player 2 has won!')
    
    elif full_board_check(board)==False:
        game_on=False
        print('Draw!')
    
    turn=switch(turn)
```
